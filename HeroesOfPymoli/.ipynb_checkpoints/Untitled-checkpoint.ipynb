```python
#import Dependencies
import pandas as pd
import numpy as np
%load_ext autoreload
```

```python
#read purchasing data file
data_df =pd.read_csv("purchase_data.csv")

data_df.head(10)
```

```python
#Display the total number of players
num_players = len(data_df["SN"].value_counts())
playercount= pd.DataFrame({"total players":[num_players]})
playercount
```

```python
#Purchasing Analysis Total 
#Number of unique items
num_items = len(data_df["Item ID"].unique())

#Avergae purchase price
ave_price = data_df["Price"].mean()

#Total number of purchase
total_purchase = (data_df["Purchase ID"].count())

#Total Revenue

total_revenue = (data_df["Price"].sum())

#Create dataframe for purchase analysis
purchase_analysis =pd.DataFrame({
    "Number of Unique Items":[num_items],
    "Average Price":'${:,.2f}'.format(ave_price),
    "Number of Purchase":[total_purchase],
    "Total Revenue":'${:,.2f}'.format(total_revenue)
    
})
purchase_analysis
```

```python
#Gender Demographics
data_df = pd.read_csv("./purchase_data.csv")
gender= data_df.groupby("Gender")
gender_total=gender.nunique()["SN"]
percentage_players= gender_total/num_players*100
gender_dataframe = pd.DataFrame ({'Total Count':gender_total,
                                 'Percentage of Players':percentage_players})
gender_dataframe['Percentage of Players']=gender_dataframe['Percentage of Players'].map('{:.2f}%'.format)
gender_dataframe.sort_values(["Total Count"],ascending=False)

```

```python
#Purchasing Analysis (Gender)
pd.options.display.float_format = '${:,.2f}'.format
gender=data_df.groupby("Gender")
gender_total=gender.nunique()["SN"]
purchase_count=gender["Purchase ID"].count()
average_purchase=gender["Price"].mean()
total_purchase=gender["Price"].sum()
average_per_person=total_purchase/gender_total

purchasing_df= pd.DataFrame({
            "Purchase Count":purchase_count,
            "Average Purchase Price":average_purchase,
            "Total Purchase Value":total_purchase,
            "Avg Total Purchase per Person":average_per_person

})
purchasing_df
            


```

```python
#Age Demographic
                    
# new dataframe
age_groups = data_df[['Age','SN']].drop_duplicates()

# creating bins by 4 years
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 80]

# bin names
age_label = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

# split players into bins 
age_groups['Age Ranges'] = pd.cut(age_groups['Age'], age_bins, labels=age_label)

# total count column values
age_count = age_groups['Age Ranges'].value_counts()


#Age Percentage
age_percentages =(age_count/num_players*100).round(2)
data_ages = pd.DataFrame ({
                         "Total Count":age_count,
                         "Percentage of Players":age_percentages.map('{:.2f}%'.format)
                        })
data_ages.sort_index()
```

```python
#Purchasing Analysis Age
#Bin the purchase_data frame by age
# age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 80]

# bin names
# age_label = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

# split players into bins 
data_df['Age Ranges'] = pd.cut(age_groups['Age'], age_bins, labels=age_label)
data_df.head(3)

#Calculate the purchase count
purchase_count = data_df.groupby(['Age Ranges']).count()['Purchase ID']

#Calculate average purchase price
mean_purchase = data_df.groupby(['Age Ranges']).mean()['Price']

#Calculate the total purchase
total_purchase_sum = data_df.groupby(['Age Ranges']).sum()['Price']

#Calculate the average purchase per person
average_pperson = total_purchase_sum/age_count

average_purchase_age = pd.DataFrame({
  
    "Purchase Count":purchase_count,
    "Average Purchase Price":mean_purchase,
    "Total Purchase Value":total_purchase_sum,
    "Avg Total Purchase per Person":average_pperson
})
average_purchase_age
```

```python
#Top Spenders
#Sort top 5 spenders
spender_statistics=data_df.groupby('SN')

#Total purchase count
spender_purchase_count=spender_statistics['Purchase ID'].count()

#calculate the avergae by name 
ave_pprice_spender= spender_statistics['Price'].mean()

#Toatl Purchase Value
total_pv_spenders= spender_statistics['Price'].sum()

#Display the dataframe
top_five_spenders = pd.DataFrame({
    "Purchase Count":spender_purchase_count,
    "Average Purchase Price":ave_pprice_spender,
    "Total Purchase Value":total_pv_spenders})
top_five_spenders

#Sort in descending order top 5 names
spenders_format_data = top_five_spenders.sort_values(["Total Purchase Value"], ascending=False).head()
spenders_format_data 
```

```python
#Most Popular
#Retrive items

popular_items = data_df[['Item ID','Item Name','Price']]



#Groupby the item id and item name
popular_groups= popular_items.groupby(['Item ID','Item Name'])

#Caluclate purchase count



popular_count= popular_groups ['Price'].count()

#total purchase value

total_purchase_value= popular_groups['Price'].sum()


#Average average item price
average_item_price = total_purchase_value/popular_count


popular_items= pd.DataFrame({
    "Purchase Count":popular_count,
    "Item Price":average_item_price,
    "Total Purchase Value":total_purchase_value
})


popular_items_sorted = popular_items.sort_values(['Purchase Count'], ascending= False).head(5)

popular_items_sorted
```

```python
#Most Profitable Items

#Sort the above table by total purchase value in descending order
popular_items_sorted = popular_items.sort_values(['Total Purchase Value'], ascending= False).head(5)
popular_items_sorted
```
